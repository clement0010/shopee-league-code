```python
import numpy as np
import pandas as pd
```

```python
shopee = pd.read_csv('order_brush_order.csv',index_col=0)
```

```python
ori_shop =shopee['shopid'].value_counts().to_dict()
ori_shop
```

```python
shopee.describe()
```

```python
filter=shopee['shopid'].value_counts()>3
filter =filter.to_dict()
filter
```

```python
sol={}
for key,value in ori_shop.items():
    if filter[key]== True:
        sol[key] = value
    else:
        sol[key] = 0
sol
```

```python
df = pd.DataFrame.from_dict(sol,orient="index")
df.to_csv('solution1.csv')
```
